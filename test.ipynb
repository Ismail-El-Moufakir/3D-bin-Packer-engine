```python
import pandas as pd
from geometry import *
from models import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from solver import *
```

```python
item_df = pd.read_csv("sample_items.csv")
item_df.head()
```

```python
items = []
for index, row in item_df.iterrows():
    quantity = row['quantity']
    if quantity > 1:
       for q in range(quantity):
           item = Item(
               id=row['id'],
               dim=Vector3(row['width'], row['length'], row['height']),
               weight=row['weight'],
               order=row['order'],
               rotatable=row['rotatable'],

           )
           items.append(item)
    else:
        item = Item(
            id=row['id'],
            dim=Vector3(row['width'], row['length'], row['height']),
            weight=row['weight'],
            order=row['order'],
            rotatable=row['rotatable'],

        )
        items.append(item)
print(f"Total items: {len(items)}")
bins = [Bin(id=i, max_weight=100, W=4, L=4, H=4) for i in range(2)]
```

```python
sol = solver(bins, items)
print(f"Total items: {len(sol.items)}")
sol.solve()
print(f"Total items packed: {len(sol.placements)}")
bin_1_placement = [p for p in sol.placements if p.bin_id == 0]
bin_2_placement = [p for p in sol.placements if p.bin_id == 1]
print(f"Bin 1: {len(bin_1_placement)} items, Bin 2: {len(bin_2_placement)} items")
```
